```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('Dataset2_UsedCarData.xlsx')

df.head()
```

```python
backup = df.copy
```

```python
df['back_legroom'] = df['back_legroom'].str.replace(' in', '')
```

```python
df.head()
```

```python
df['body_type'] = df['body_type'].replace('SUV / Crossover', 'SUV')
df = df.dropna(subset=['body_type'])
df['body_type'].unique()
```

```python
df = df.drop('combine_fuel_economy', axis = 1)

mean_city_fuel_economy = df['city_fuel_economy'].mean()
df['city_fuel_economy'].fillna(mean_city_fuel_economy, inplace=True)

df = df.drop('engine_cylinders', axis = 1)

df['engine_type'] = df['engine_type'].str.replace('V6 Diesel', 'V6D')
df['engine_type'] = df['engine_type'].str.replace('V8 Biodiesel', 'V8B')
df['engine_type'] = df['engine_type'].str.replace('V8 Flex Fuel Vehicle', 'V8FFV')
df['engine_type'] = df['engine_type'].str.replace('V6 Flex Fuel Vehicle', 'V6FFV')
df['engine_type'] = df['engine_type'].str.replace('I4 Hybrid', 'I4H')
df['engine_type'] = df['engine_type'].str.replace('V6 Hybrid', 'V6H')
df['engine_type'] = df['engine_type'].str.replace('I4 Flex Fuel Vehicle', 'I4FFV')
df['engine_type'] = df['engine_type'].str.replace('I4 Diesel', 'I4D')

mode_engine_type = df['engine_type'].mode()[0]
df['engine_type'] = df['engine_type'].fillna(mode_engine_type)

threshold = 25
counts = df['exterior_color'].value_counts()
df['exterior_color'] = np.where(df['exterior_color'].isin(counts.index[counts < threshold]), 'Other', df['exterior_color'])
mode_exterior_color = df['exterior_color'].mode()[0]
df['exterior_color'] = df['exterior_color'].fillna(mode_exterior_color)


mode_fleet = df['fleet'].mode()[0]
df['fleet'].fillna(mode_fleet, inplace=True)
df['fleet'] = df['fleet'].map({1.0: 1, 0.0: 0})

df = df.drop('frame_damaged', axis=1)

df['franchise_dealer'] = df['franchise_dealer'].map({True: 1, False: 0})


df['front_legroom'] = df['front_legroom'].str.replace(' in', '')
df['fuel_tank_volume'] = df['fuel_tank_volume'].str.replace(' gal', '')


mode_franchise_make = df['franchise_make'].mode()[0]
df['franchise_make'] = df['franchise_make'].fillna(mode_franchise_make)

mode_has_accidents = df['has_accidents'].mode()[0]
df['has_accidents'].fillna(mode_has_accidents, inplace=True)
df['has_accidents'] = df['has_accidents'].map({1.0: 1, 0.0: 0 })

df['height'] = df['height'].str.replace(' in', '')

mean_highway_fuel_economy = df['highway_fuel_economy'].mean()
df['highway_fuel_economy'].fillna(mean_highway_fuel_economy, inplace=True)

mean_horsepower = df['horsepower'].mean()
df['horsepower'].fillna(mean_horsepower, inplace=True)

threshold = 25
counts = df['interior_color'].value_counts()
df['interior_color'] = np.where(df['interior_color'].isin(counts.index[counts < threshold]), 'Other', df['interior_color'])
mode_interior_color = df['interior_color'].mode()[0]
df['interior_color'] = df['interior_color'].fillna(mode_interior_color)

mode_isCab = df['isCab'].mode()[0]
df['isCab'].fillna(mode_isCab, inplace=True)
df['isCab'] = df['isCab'].map({1.0: 1, 0.0: 0 })

df = df.drop('is_certified', axis = 1)
df = df.drop('is_cpo', axis = 1)

df['is_new'] = df['is_new'].map({True: 1, False: 0})

df = df.drop('is_oemcpo', axis = 1)

df['length'] = df['length'].str.replace(' in', '')

df['listing_color'] = df['listing_color'].replace('UNKNOWN', pd.NA)
mode_listing_color = df['listing_color'].mode()[0]
df['listing_color'] = df['listing_color'].fillna(mode_listing_color)

df['maximum_seating'] = df['maximum_seating'].str.replace(' seats', '')

df['mileage'].replace(0, df['mileage'].mean(), inplace=True)
df['mileage'].fillna(df['mileage'].mean(), inplace=True)

mode_salvage = df['salvage'].mode()[0]
df['salvage'].fillna(mode_salvage)
df['salvage'] = df['salvage'].map({1.0: 1, 0.0: 0})

mean_seller_rating = df['seller_rating'].mean()
df['seller_rating'].fillna(mean_seller_rating, inplace=True)


mode_transmission = df['transmission'].mode()[0]
df['transmission'] = df['transmission'].fillna(mode_transmission)


print(df['transmission_display'].unique())
mode_transmission_display = df['transmission_display'].mode()[0]
df['transmission_display'] = df['transmission_display'].fillna(mode_transmission_display)

df = df.drop('vehicle_damage_category', axis = 1)

mode_wheel_system = df['wheel_system'].mode()[0]
df['wheel_system'] = df['wheel_system'].fillna(mode_wheel_system)

df = df.drop('wheel_system_display', axis = 1)

df['wheelbase'] = df['wheelbase'].str.replace(' in', '')
df['width'] = df['width'].str.replace(' in', '')

```

```python
df['back_legroom'] = pd.to_numeric(df['back_legroom'], errors='coerce')
df['back_legroom'] = df['back_legroom'].replace(pd.NA, df['back_legroom'].mean())
df['back_legroom'] = df['back_legroom'].fillna(df['back_legroom'].mean())

df['engine_displacement'] = df['engine_displacement'].fillna(df['engine_displacement'].mean())

df['front_legroom'] = pd.to_numeric(df['front_legroom'], errors='coerce')
df['front_legroom'] = df['front_legroom'].replace(pd.NA, df['front_legroom'].mean())
df['front_legroom'] = df['front_legroom'].fillna(df['front_legroom'].mean())

df['fuel_tank_volume'] = pd.to_numeric(df['fuel_tank_volume'], errors='coerce')
df['fuel_tank_volume'] = df['fuel_tank_volume'].fillna(df['fuel_tank_volume'].mean())

mode_fuel_type = df['fuel_type'].mode()[0]
df['fuel_type'] = df['fuel_type'].fillna(mode_fuel_type)

df['height'] = pd.to_numeric(df['height'], errors='coerce')
df['height'] = df['height'].fillna(df['height'].mean())

df['length'] = pd.to_numeric(df['length'], errors='coerce')
df['length'] = df['length'].fillna(df['length'].mean())

df['maximum_seating'] = pd.to_numeric(df['maximum_seating'], errors='coerce')
df['maximum_seating'] = df['maximum_seating'].fillna(df['maximum_seating'].mean())

df['wheelbase'] = pd.to_numeric(df['wheelbase'], errors='coerce')
df['wheelbase'] = df['wheelbase'].fillna(df['wheelbase'].mean())

df['width'] = pd.to_numeric(df['width'], errors='coerce')
df['width'] = df['width'].fillna(df['width'].mean())

def extract_torque(torque_string):
    if pd.isna(torque_string):
        return None
    return int(torque_string.split()[0])

df['torque'] = df['torque'].apply(extract_torque)

df['torque'] = df['torque'].fillna(df['torque'].mean())

unnecessary_columns = ['main_picture_url', 'major_options', 'owner_count', 'power', 'theft_title', 'trimId', 'trim_name', 'listed_date']

for i in unnecessary_columns:
    df = df.drop(i, axis=1)
```

```python
df['fleet'] = df['fleet'].astype('object')
df['has_accidents'] = df['has_accidents'].astype('object')
df['is_new'] = df['is_new'].astype('object')

df = df.drop(columns=['model_name','salvage','sp_id','sp_name','transmission_display','vin','city',], axis=1)

df['fleet'] = df['fleet'].map({0:False, 1:True})
df['has_accidents'] = df['has_accidents'].map({0:False, 1:True})
df['is_new'] = df['is_new'].map({0:False, 1:True})

```

```python
cat_cols = ['body_type', 'engine_type', 'exterior_color', 'franchise_make',
            'fuel_type', 'interior_color', 'listing_color', 'make_name',
            'transmission', 'wheel_system']

df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)


df_encoded.select_dtypes(include=['float64']).columns
```

```python
from sklearn.preprocessing import StandardScaler

columns_to_scale = ['back_legroom', 'fuel_tank_volume', 'highway_fuel_economy',
                    'horsepower', 'maximum_seating', 'mileage', 'seller_rating']

scaler = StandardScaler()

df_encoded[columns_to_scale] = scaler.fit_transform(df_encoded[columns_to_scale])
```

```python
df_encoded.select_dtypes(include=['float64']).head()
```

```python
X = df_encoded.drop('price', axis=1)
y = df_encoded['price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
pred = y_pred.flatten()
act = y_test.values.flatten()

result = pd.DataFrame({'actual':act, 'predicted':pred})


result.head()
```

```python
import matplotlib.pyplot as plt

# Assuming 'result' is your DataFrame with columns 'actual' and 'predicted'
plt.scatter(result['actual'], result['predicted'], alpha=0.5, label='Actual vs Predicted')
plt.plot([result['actual'].min(), result['actual'].max()], [result['actual'].min(), result['actual'].max()], color='red', linestyle='--', linewidth=2, label='Regression Line')

plt.title('Actual vs Predicted Values with Regression Line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)

rf_regressor.fit(X_train, y_train)
```

```python
rf_y_pred = rf_regressor.predict(X_test)
```

```python
rf_pred = rf_y_pred.flatten()
rf_act = y_test.values.flatten()

rf_result = pd.DataFrame({'actual':rf_act, 'predicted':rf_pred})


rf_result.head(50)
```

```python
# Assuming 'result' is your DataFrame with columns 'actual' and 'predicted'
plt.scatter(rf_result['actual'], rf_result['predicted'], alpha=0.5, label='Actual vs Predicted')
plt.plot([rf_result['actual'].min(), rf_result['actual'].max()], [rf_result['actual'].min(), rf_result['actual'].max()], color='red', linestyle='--', linewidth=2, label='Regression Line')

plt.title('Actual vs Predicted Values with Regression Line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(rf_act, rf_pred)
mse = mean_squared_error(rf_act, rf_pred)
r2 = r2_score(rf_act, rf_pred)

print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
